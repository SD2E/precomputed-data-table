```python
import pandas as pd
import numpy as np
```

```python
# Read in indiviual experiment dataframes
folder = 'notebook_data/RF+OD/'
df1 = pd.read_csv(folder+'RF_OD_CEN-PK-Inducible-CRISPR-4-Day-Obstacle-Course_20200721164700.csv', index_col=0)
df2 = pd.read_csv(folder+'RF_OD_YeastSTATES-CRISPR-Short-Duration-Time-Series-20191208_20200610192131.csv', index_col=0)
df3 = pd.read_csv(folder+'RF_OD_YeastSTATES-CRISPR-Short-Duration-Time-Series-35C_20200625205807.csv', index_col=0)
df4 = pd.read_csv(folder+'RF_OD_YeastSTATES-OR-Gate-CRISPR-Dose-Response_20200625204022.csv', index_col=0)
```

```python
# Insert experiment name column
df1.insert(0, 'experiment', 'CEN-PK-Inducible-CRISPR-4-Day-Obstacle-Course_20200721164700')
df2.insert(0, 'experiment', 'YeastSTATES-CRISPR-Short-Duration-Time-Series-20191208_20200610192131')
df3.insert(0, 'experiment', 'YeastSTATES-CRISPR-Short-Duration-Time-Series-35C_20200625205807')
df4.insert(0, 'experiment', 'YeastSTATES-OR-Gate-CRISPR-Dose-Response_20200625204022')
```

```python
# Identify columns that do not exist in df4 and drop them from first three
#print(df1.columns)
#print(df2.columns)
#print(df3.columns)
#print(df4.columns)

df1_extra_cols = ['experiment_reference', 'strain_class', 'standard_type', 'control_type', 'sample_contents']
df2_extra_cols = ['Unnamed: 0.1', 'experiment_reference', 'replicate_group', 'replicate_group_string', 'strain_class']
df3_extra_cols = ['experiment_reference', 'strain_class', 'standard_type', 'control_type', 'sample_contents']

df1 = df1.drop(columns=df1_extra_cols)
df2 = df2.drop(columns=df2_extra_cols)
df3 = df3.drop(columns=df3_extra_cols)
```

```python
# All dataframes should have the same columns
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
```

```python
# Concatenate into one dataframe
all_df = pd.concat([df1, df2], ignore_index=True)
all_df = pd.concat([all_df, df3], ignore_index=True)
all_df = pd.concat([all_df, df4], ignore_index=True)
```

```python
all_df
```

```python
# Split aggregated dataframe into matches and mismatches
all_matches = all_df.loc[np.invert(all_df['dead']).astype(int) == np.round(all_df['RF_prediction_mean']).astype(int),]
all_mismatches = all_df.loc[np.invert(all_df['dead']).astype(int) != np.round(all_df['RF_prediction_mean']).astype(int),]
```
